# Практическое задание к уроку 4 

### 1. Расскажите, как работает регуляризация в решающих деревьях, какие параметры мы штрафуем в данных алгоритмах?


В процессе построения дерева решений могут создаваться слишком сложные конструкции, которые будут полностью подстроены под тренировочные данные но бесполезны на тестовых данных. Для того, чтобы её избежать, необходимо использовать методы «регулирования глубины дерева». При использования ансамбля деревьев решений при обучении каждому голосующему дереву присваивается вес. для избежания слишком больших весов для одного участника накладываются штрафы за большие веса

### 2. По какому принципу рассчитывается "важность признака (feature_importance)" в ансамблях деревьев?

При построении случайного леса вначале генерируется количество бутстрап-выборок, равное количеству деревьев в алгоритме. Для уменьшения корреляции базовых алгоритмов рандомизируют сам процесс построения каждого дерева: в каждой вершине $j$-й признак выбирается не из всего пространства признаков, а из его случайного подмножества размера $m$, которое каждый раз выбирается заново и вычисляем суммарное уменьшение ошибки с помощью ветвлений по каждому признаку.
